```python
import pandas as pd

# Initialize empty lists to hold different kinds of data
general_flight_info = []
segment_info = []
cabin_price_info = []
cabin_pricing_per_flight = []

# Iterate through each flight in the 'outboundFlights' dictionary
for flight_key, flight_value in data['outboundFlights'].items():
    # Extracting general flight information
    general_flight_info.append({
        'FlightKey': flight_key,
        'Origin': flight_value.get('origin', {}).get('code', None),
        'Destination': flight_value.get('destination', {}).get('code', None),
        'StartTimeInLocal': flight_value.get('startTimeInLocal', None),
        'EndTimeInLocal': flight_value.get('endTimeInLocal', None),
        'ConnectionDuration': flight_value.get('connectionDuration', None),
        'Stops': flight_value.get('stops', None)
    })

    # Extracting segment information
    for segment in flight_value.get('segments', []):
        segment_info.append({
            'FlightKey': flight_key,
            'SegmentID': segment.get('id', None),
            'DepartureAirport': segment.get('departureAirport', {}).get('code', None),
            'ArrivalAirport': segment.get('arrivalAirport', {}).get('code', None),
            'DepartureTimeInLocal': segment.get('departureDateTimeInLocal', None),
            'ArrivalTimeInLocal': segment.get('arrivalDateTimeInLocal', None),
            'Duration': segment.get('duration', None),
            'FlightNumber': segment.get('flightNumber', None),
        })

    # Extracting cabin and price information
    for cabin_type, cabin_details in flight_value.get('cabins', {}).items():
        for product_type, product_details in cabin_details.items():
            for product_code, product_info in product_details.get('products', {}).items():
                cabin_price_info.append({
                    'FlightKey': flight_key,
                    'CabinType': cabin_type,
                    'ProductType': product_type,
                    'ProductCode': product_code,
                    'ProductName': product_info.get('productName', None),
                    'TotalPrice': product_info.get('price', {}).get('totalPrice', None),
                    'Currency': product_info.get('price', {}).get('currency', None)
                })

    # Extracting cabin and price information for each outbound flight
    for cabin_type, cabin_details in flight_value.get('cabins', {}).items():
        for product_type, product_details in cabin_details.items():
            for product_code, product_info in product_details.get('products', {}).items():
                cabin_pricing_per_flight.append({
                    'FlightKey': flight_key,
                    'CabinType': cabin_type,
                    'ProductType': product_type,
                    'ProductCode': product_code,
                    'ProductName': product_info.get('productName', None),
                    'TotalPrice': product_info.get('price', {}).get('totalPrice', None),
                    'Currency': product_info.get('price', {}).get('currency', None),
                    'Points': product_info.get('price', {}).get('points', None),
                    'PointsAfterDiscount': product_info.get('price', {}).get('pointsAfterDiscount', None)
                })

# Create Pandas DataFrames from the lists
general_flight_info_df = pd.DataFrame(general_flight_info)
segment_info_df = pd.DataFrame(segment_info)
cabin_price_info_df = pd.DataFrame(cabin_price_info)
cabin_pricing_per_flight_df = pd.DataFrame(cabin_pricing_per_flight)

```
